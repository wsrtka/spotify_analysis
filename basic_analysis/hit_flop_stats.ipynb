```python
from database.load_data import load_sql
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
historical_data = ["dataset_60s",
        "dataset_70s",
        "dataset_80s",
        "dataset_90s",
        "dataset_00s",
        "dataset_10s"]
spotify_dataset = 'spotify_dataset'
```

## 60s

```python
dataset_60s = load_sql('dataset_60s')
```

```python
dataset_60s.columns
```

```python
cols = dataset_60s.columns[3:-1]
```

```python
dataset_60s_hit = dataset_60s.loc[dataset_60s.target == 1]
dataset_60s_flop = dataset_60s.loc[dataset_60s.target == 0]
```

```python
plt.figure(figsize=(19, 25))
for idx, col in enumerate(cols):
    d_hit = dataset_60s_hit[col]
    d_flop = dataset_60s_flop[col]
    plt.subplot(len(cols), 1, idx+1)
    plt.title(col)
    plt.hist(d_hit, label='hit')
    plt.hist(d_flop, label='flop', alpha = 0.5)
    plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(19, 8))
pca = PCA()
scaler = StandardScaler()
d = dataset_60s.iloc[:, 3:-1].values
d = scaler.fit_transform(d)
pca.fit(d)
dataset_60s_hit_pca = pca.transform(scaler.transform(dataset_60s_hit.iloc[:, 3:-1].values))
dataset_60s_flop_pca = pca.transform(scaler.transform(dataset_60s_flop.iloc[:, 3:-1].values))
plt.scatter(dataset_60s_hit_pca[:, 0], dataset_60s_hit_pca[:, 1], label='hit')
plt.scatter(dataset_60s_flop_pca[:, 0], dataset_60s_flop_pca[:, 1], label='flop')
plt.legend()
```

## spotify dataset

```python
spotify_dataset = load_sql('spotify_dataset')
```

```python
spotify_dataset
```

```python
genres = set()

for g in spotify_dataset.Genre:
    g = eval(g.strip() or '[]')
    
    for genre in g:
        genres.add(genre)
genres
```

```python
music_genres = list(genres)
```

```python
spotify_dataset.values.shape
```

```python
new_features = np.zeros((spotify_dataset.values.shape[0], len(music_genres)))
```

```python

for ex_id, g in enumerate(spotify_dataset.Genre):
    g = set(eval(g.strip() or '[]'))
    for idx, genra in enumerate(music_genres):
        if genra in g:
            new_features[ex_id, idx] = 1
```

```python
features = pd.DataFrame(new_features, columns=music_genres, index=spotify_dataset.index)
features
```

```python
new_spotify_df = spotify_dataset.loc[:, spotify_dataset.columns!='Genre']
new_spotify_df = pd.concat((new_spotify_df, features), axis=1)
new_spotify_df.sort_values('Number of Times Charted', ascending=False)
```
